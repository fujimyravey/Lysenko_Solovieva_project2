```python
crime = df['Description'].value_counts()

plt.figure(figsize = (14, 8))
crime.plot(kind = 'bar', color = 'red')
plt.title('Частота разных типов зарегистрированных преступлений в Балтиморе с 2012 по 2017',  fontsize = 20)
plt.xlabel('Тип преступления',  fontsize = 16)
plt.ylabel('Количество зарегистрированных преступлений',  fontsize = 16)
plt.xticks(rotation = 50)
plt.show()
```

```python
weapon = df['Weapon'].value_counts()

plt.figure(figsize = (14, 8))
weapon.plot(kind = 'bar', color = 'purple')
plt.title('Преступления совершаемые с использованием оружия',  fontsize = 20)
plt.xlabel('Тип оружия',  fontsize = 16)
plt.ylabel('Количество преступлений',  fontsize = 16)
plt.xticks(rotation=45)
plt.tight_layout()  
plt.show()
```

# Часть 2. Визуализация распределения преступлений двух типов на карте

```python
!pip install folium
import folium 
from folium.plugins import HeatMap
```

```python
Robbery=df.loc[df.Description=='ROBBERY - STREET'][['Latitude','Longitude']]
Robbery.Latitude.fillna(0, inplace = True)
Robbery.Longitude.fillna(0, inplace = True) 

BaltimorMap=folium.Map(location=[39.290197, -76.612693],zoom_start=11)
HeatMap(data=Robbery, radius=16).add_to(BaltimorMap)

BaltimorMap
```

## Данная карта показывает интенсивность ограблений, совершенных на улицах Балтимора с 2012 по 2017 г.

С помощью данной визуализации мы хотели выявить наиболее опасные и неблагополучные районы города, но в силу большой частоты уличных ограблений, как преступного явления, без дополнительных данных о других видах преступности, невозможно сделать однозначный вывод о совокупном уровне криминала в том или ином районе города.

```python
!pip install folium
import folium 
from folium.plugins import HeatMap
import pandas as pd

df = pd.read_csv(r'C:\Users\yana3\OneDrive\Desktop\python\project_2\BPD.csv', sep=';', on_bad_lines='skip')

df.head(10)

Shooting=df.loc[df.Description=='SHOOTING'][['Latitude','Longitude']]
Shooting.Latitude.fillna(0, inplace = True)
Shooting.Longitude.fillna(0, inplace = True) 

BaltimorMap=folium.Map(location=[39.290197, -76.612693],zoom_start=11)
HeatMap(data=Shooting, radius=16).add_to(BaltimorMap)

BaltimorMap
```

## Данная карта показывает интенсивность стрельбы, совершенной на улицах Балтимора с 2012 по 2017 г.

Стрельба, как тип преступления — наиболее яркий маркер, указывающий на низкий уровень благополучия того или иного района. Данные о таком типе преступления хорошо дополняют предыдущую визуализацию.
